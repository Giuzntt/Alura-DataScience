# Tratamento de dados faltantes

```python
import pandas as pd    
```

```python
dados =  pd.read_csv('dados_residencial.csv', sep=';')
dados.head(10)
```

```python
dados.isnull() # Verifica se existe dados faltantes
dados.notnull() # Verifica se existe dados faltantes
```

```python
dados.info() # Informações sobre o dataframe e seus tipos de dados
```

```python
dados[dados['Valor'].isnull()] # Verifica se existe dados faltantes
```

```python
A = dados.shape[0] # Quantidade de linhas
dados.dropna(subset=['Valor'], inplace=True) # Remove os dados faltantes'
B = dados.shape[0] # Quantidade de linhas
A-B
```

```python
dados[dados['Valor'].isnull()]
```

```python
# Tratamento de dados faltantes
dados[dados['Condominio'].isnull()].shape[0]
```

```python
selecao = (dados['Tipo']== 'Apartamento') & (dados['Condominio'].isnull())

A = dados.shape[0] # Quantidade de linhas
dados = dados[~selecao]
B = dados.shape[0] # Quantidade de linhas
A-B
```

```python
dados[dados['Condominio'].isnull()].shape[0]
```

```python
dados.fillna({'Condominio': 0, 'IPTU': 0})
# O que o inplace faz?

# 
```
